```python
#Al ver que hay datos atpicos y que la media no esta bien representada, decido eliminarlos de la base de datos.
stats = datos.describe().transpose()
stats['IQR'] = stats['75%'] - stats['25%']
stats
```

```python
outliers = pd.DataFrame(columns=datos.columns)

for col in stats.index:
    iqr = stats.at[col,'IQR']
    cutoff = iqr * 1.5
    lower = stats.at[col,'25%'] - cutoff
    upper = stats.at[col,'75%'] + cutoff
    results = datos[(datos[col] < lower) | 
                   (datos[col] > upper)].copy()
    results['Outlier'] = col
    outliers = outliers.append(results)
```

```python
outliers.shape
```

```python
outliers.Outlier.value_counts()
```

```python
datos_atipico=(outliers[outliers["Outlier"]].index)
```

```python
datos.drop(datos_atipico,axis=0,inplace=True)
```

```python
#Ahora creo unas funciones para pasar las variables cualitativas a tipo numericas.
```

```python
def talla (e):
    if e=="Ideal":
        return 0.95
    elif e=="Premium":
        return 1
    elif e=="Very Good":
        return 0.8
    elif e=="Good":
        return 0.5
    elif e=="Fair":
        return 0
    else:
        pass
    return e
```

```python
datos["cut"]=datos["cut"].apply(talla)
```

```python
def color (e):
    if e=="D":
        return 0
    elif e=="E":
        return 0.85
    elif e=="F":
        return 0.91
    elif e=="G":
        return 0.95
    elif e=="H":
        return 1
    elif e=="I":
        return 0.91
    elif e=="J":
        return 0.1
    else:
        pass
    return e
```

```python
datos["color"]=datos["color"].apply(color)
```

```python
def claridad (e):
    if e=="IF":
        return 0
    elif e=="VVS1":
        return 0.3
    elif e=="VVS2":
        return 0.78
    elif e=="VS1":
        return 0.87
    elif e=="VS2":
        return 0.9
    elif e=="SI1":
        return 0.25
    elif e=="SI2":
        return 1
    elif e=="I1":
        return 0.1
    else:
        pass
    return e
```

```python
datos["clarity"]=datos["clarity"].apply(claridad)
```
